```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import BernoulliNB
pets_df = pd.read_csv("/content/pets.csv")
pets_df
```

```python
x=pets_df.drop(['Can we pet them'],axis=1)
x

y=pets_df['Can we pet them']
y
```

```python
#Creating ML Model
BerNB=BernoulliNB()
#Fitting our ML Model
one_hot_data = pd.get_dummies(x)
pets = BerNB.fit(one_hot_data,y)
one_hot_data
```

```python
# Prediction
# Animal = Cow, Size = Big, Body Color = White
prediction = pets.predict([[0, 0, 1, 0, 1, 0, 0, 1, 0]])
prediction
```
