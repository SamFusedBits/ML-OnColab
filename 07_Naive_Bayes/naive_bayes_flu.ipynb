**Naive Bayes Code**

```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import BernoulliNB
flu_df = pd.read_csv("/content/flu.csv")
flu_df
```

```python
x=flu_df.drop(['Flue'],axis=1)
x
```

```python
y=flu_df['Flue']
y
```

```python
#Creating ML Model
BerNB=BernoulliNB()
#Fitting our ML Model
one_hot_data = pd.get_dummies(x)
train_df = BerNB.fit(one_hot_data,y)
one_hot_data
```

Refer the concept behind One-Hot Encoding -
One hot encoding is a technique that we use to represent categorical variables as numerical values in a machine learning model.

```python
# Prediction
#Chills = T, RunnyNose = N, Headache = Mild, Fever = Y
prediction_y = train_df.predict([[0,1,1,0,1,0,0,0,1]])
prediction_y
```

```python
#Chills = N, RunnYNose = N, Headache = No, Fever = N    - Whichever column's are mentioned put values in their place as 1 else 0 for remaining columns
prediction_n = train_df.predict([[1,0,1,0,0,1,0,1,0]])
prediction_n
```
