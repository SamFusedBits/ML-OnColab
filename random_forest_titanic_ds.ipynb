## **Apply AdaBoost Algorithm on Titanic Dataset**

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df1 = pd.read_csv(url)
```

```python
df1 = pd.DataFrame(df1)
df1
newdf = df1.interpolate()
newdf
df1 = newdf.fillna({'Cabin':'Not found'})
df1
```

```python
x = df1.drop(['Survived','PassengerId'],axis=1)
y = df1['Survived']
x
```

```python
y
```

```python
one_hot_data = pd.get_dummies(x)
one_hot_data
```

```python
#Splitting the dataset into training and test samples
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(one_hot_data,y,test_size=0.3)
```

```python
#Initializing Adaboost classifier and fitting the training data
from sklearn.ensemble import AdaBoostClassifier
```

```python
#Create adaboost classifer object
abc = AdaBoostClassifier(n_estimators=50,learning_rate=1,random_state=42)
abc
```

```python
#Train Adaboost Classifier
model1 = abc.fit(x_train,y_train)
model1
```

```python
y_pred = model1.predict(x_test)
y_pred
```

```python
#Import skikit-learn metrics module for accuracy calculation
from sklearn.metrics import accuracy_score

print("AdaBoost Classifier Model Accuracy: ",accuracy_score(y_test,y_pred))
```

# **Decision Tree Classifier**

```python
#Apply Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf


#Train the model
clf = clf.fit(x_train,y_train)
clf

#To check the accuracy
y_pred = clf.predict(x_test)

#Import skikit-learn metrics module for accuracy calculation
from sklearn.metrics import accuracy_score

print("AdaBoost Classifier Model Accuracy: ",accuracy_score(y_test,y_pred))
```
