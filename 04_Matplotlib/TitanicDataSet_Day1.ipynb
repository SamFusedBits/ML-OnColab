```python
def impute_age(cols):
  Age=cols[0]
  Pclass=cols[1]
  if pd.isnull(Age):
    if Pclass==1:
      return 38
    elif Pclass==2:
      return 30
    else:
      return 25
  else:
      return Age

titanic['Age']=titanic[['Age','Pclass']].apply(impute_age,axis=1)  #Here axis=1 specifies column
```

```python
# d. For particular attributes use different attributes of line graph
import pandas as pd
import matplotlib.pyplot as plt
url = '/content/Assignement_Data2.csv'
data_2 = pd.read_csv(url)
df = pd.DataFrame(data_2)
df['col_1'].plot(kind='line', figsize=(10, 6), linestyle='--', marker='o', color='red', label='col_1')
df['col_2'].plot(kind='line', linestyle='-.', marker='s', color='blue', label='col_2')
plt.title('Line Graph')
plt.xlabel('Index')
plt.ylabel('Values')
```

```python
# e. Area Plot for whole dataset & for any of the Column/feature
df.plot(kind='area', figsize=(10, 6))
plt.title('Area Plot for Whole Dataset')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend(title='Columns')
plt.tight_layout()
plt.show()

df['col_3'].plot(kind='area', figsize=(10, 6), color='blue')
plt.title('Area Plot for col_3')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend(['col_3'])
plt.tight_layout()
plt.show()
```

```python
# f. Show legend
df.plot(kind='line', figsize=(10, 6))
plt.title('Line Graph with Legend')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend(title='Columns', loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()
```

```python
# g. Move it to right and up
df.plot(kind='line', figsize=(10, 6))
plt.title('Line Graph - Moved to Right and Up')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend(title='Columns', loc='upper right', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```
