```python
#Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. Matplotlib makes easy things easy and hard things possible.
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[40,41,42,38,37,39,36]
plt.plot(x,y)
```

```python
#Reference - https://matplotlib.org/stable/api/index.html
plt.plot(x,y,color="Red")
```

```python
plt.plot(x,y,linestyle=":",linewidth=4,color="Green")  #linestyle or ls{'-', '--', '-.', ':', '', (offset, on-off-seq), ...}
```

```python
#Add a title Day on x-axis, Temperature on y-axis and title 'Weather'
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,linestyle="--",linewidth=5,color="Orange")
```

```python
#The matplotlib.markers module provides a collection of marker styles used for data visualization in Matplotlib plots.
#These markers represent individual data points, enhancing clarity and distinction within graphs and charts.
plt.plot(x,y,color="blue",marker="*",linestyle='')
```

```python
#Reveneus for the major US tech companies (Bar Chart)
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
plt.bar(company,revenue)
plt.ylabel("Revenue(bln)")
plt.title("US Tech Gaint Stocks")
```

```python
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar)                                         #This method uses numpy.histogram to bin the data in x and count the number of values in each bin, then draws the distribution either as a BarContainer or Polygon.
```

```python
plt.hist(blood_sugar,bins=3)    #Continuous distribution
```

```python
plt.hist(blood_sugar,bins=3,rwidth=0.95)  #1<Relative width(rwidth) < 0   #Discrete distribution
```

```python
#We can define the range of bins like this
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95)
```

```python
#Blood Sugar of Women
blood_sugar_women=[67,98,89,120,133,150,84,49,89,79,129,112,100]
plt.hist([blood_sugar,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95)
```

```python
blood_sugar_women=[67,98,89,120,133,150,84,49,89,79,129,112,100]
plt.hist([blood_sugar,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,label=['Men','Women'],orientation="horizontal")     #Labeling for specifying Men and Women Diabetic Levels & changing the orientation to horizontal
plt.legend()
plt.xlabel('Sugar Range')                            #X-axis
plt.ylabel('Total Number of Patients')               #Y-axis
plt.title('Blood Sugar Analysis')                    #Title
```
