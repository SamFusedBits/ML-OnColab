```python
# Assignment
# Create a structure on the Titanic Dataset using subplot, incorporting the six mentioned types as illustrated.
# Apply:
# Line Plot
# Bar Chart
# Scatter Plot
# Area Plot
# Pie Chart
# Histogram

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
url="https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url)
# Creating subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

# Line Plot
sns.lineplot(x='Age', y='Fare', data=titanic, ax=axes[0, 0])
axes[0, 0].set_title('Line Plot')

# Bar Chart
sns.barplot(x='Pclass', y='Survived', data=titanic, ax=axes[0, 1])
axes[0, 1].set_title('Bar Chart')

# Scatter Plot
sns.scatterplot(x='Age', y='Fare', data=titanic, hue='Survived', ax=axes[0, 2])
axes[0, 2].set_title('Scatter Plot')

# Area Plot
sns.kdeplot(data=titanic, x='Age', fill=True, ax=axes[1, 0])
axes[1, 0].set_title('Area Plot')

# Pie Chart
class_counts = titanic['Pclass'].value_counts()
axes[1, 1].pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90)
axes[1, 1].set_title('Pie Chart')

# Histogram
sns.histplot(titanic['Age'], bins=20, kde=True, ax=axes[1, 2])
axes[1, 2].set_title('Histogram')

# Adjust layout
plt.tight_layout()
```
