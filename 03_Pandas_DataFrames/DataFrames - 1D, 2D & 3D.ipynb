```python
#Series 1-D
#Data Frames 2-D (Widely used) #Eg: DataFrame is a container of series
#Panel 3-D #Eg: Panel is a container of DataFrame
#A series can be created using using various inputs like -
#Scalar value or constant
import pandas as pd
import numpy as np
s1=pd.Series(5)
print(s1)
```

```python
s1=pd.Series(5,index=[0,1,2,3])
print(s1)
```

```python
#Array
data=np.array([33,44,55,66])
s2=pd.Series(data)
print(s2)
```

```python
#Create a Series from dict
dict={'a':111,'b':222,'c':333}
s3=pd.Series(dict)
print(s3)
```

```python
data={'p':101,'q':102,'r':103,'s':104}
s4=pd.Series(data)
print(s4)
```

```python
#Data Frame is a 2-D data structure #Similar to spreadsheet
#Used to manage large and complex data in tabular format. #It has row and column indexes
#axis = 0 refers to row index
#axis = 1 refers to column index

#Creating Dataframe
#A pandas dataFrame can be created using the following constructor
#- pandas.DataFrame(data, index, columns, dtype)

#Data - Values to be passed in DataFrame
#Index - Used to label index for rows
#Columns - Used to label index for columns
#Dtype - Used to define datatype for columns
```

```python
#DataFrame can accept datafrom
#List Dictionaary Tuple String Series Another dataframe
```

```python
#E.g. 1- DataFrame with default indexes
apple=[10,20,30,40]
banana=[11,22,33,44]
dff=pd.DataFrame([apple,banana])
print(dff)
```

```python
#E.g. 2- DataFrame with labelled indexes
dff=pd.DataFrame([apple,banana], index=['apple','banana'],columns=['jan','feb','mar','apr'])
print(dff)
```

```python
#E.g. 3- Create DataFrame using multiple list of different length
apple=[10,20,30,40]
banana=[11,22,33,44]
orange=[55,66,77,88,99]
rw=['apple','banana','orange']
cl=['jan','feb','mar','apr','may']

dff=pd.DataFrame([apple,banana,orange],index=rw,columns=cl)
print(dff)
#NaN is automatically inserted if no matching values is found for columns
```

```python
#4 Creating DataFrame using Series
s1=pd.Series([10,20,30],index=[1,2,3])
df2=pd.DataFrame(s1)
df2
```

```python
#5 Creating DataFrame using Multiple Series
s2=pd.Series([100,200,300],index=[1,2,3])
df3=pd.DataFrame([s1,s2])
df3
```

```python
s1=pd.Series([10,20,30],index=['jan','feb','mar'])
s2=pd.Series([100,200,300],index=['jan','feb','mar'])
df3=pd.DataFrame([s1,s2])
df3
```

```python
df5=pd.DataFrame([s1,s2],index=['Sale1','Sale2'])
df5
```

```python
#7 Create dataframe passing dictionary using key value pair
s3=pd.Series(['kabaddi','kho-kho','volley ball'])
s4=pd.Series([7,12,5])

dict={'sports':s3,'No of Players:':s4}
df5=pd.DataFrame(dict)
df5
```

```python
df5=pd.DataFrame([s4,s3],index=['R1','R2'])
df5
```
