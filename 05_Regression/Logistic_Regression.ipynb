```python
#@title Logistic Regression - It is one of the technique used for Binary Classification.
#Here the dependent variable is binary. Which is either TRUE or FALSE / 1 or 0 / YES or NO

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

url='/content/SampleLogitReg - SampleLogitReg.csv'
data = pd.read_csv(url)
data
```

```python
# Extracting features 'Age' and 'EstimatedSalary' and target variable 'Purchased' from the DataFrame 'data'.
x=data[['Age','EstimatedSalary']]
y=data['Purchased']

# Create a 1x2 subplot for side-by-side scatter plots.
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

# Scatter plot for Age vs Purchased
axs[0].scatter(data['Age'], y)
axs[0].set_title('Age vs Purchased')
axs[0].set_xlabel('Age')
axs[0].set_ylabel('Purchased')

# Scatter plot for EstimatedSalary vs Purchased
axs[1].scatter(data['EstimatedSalary'], y)
axs[1].set_title('EstimatedSalary vs Purchased')
axs[1].set_xlabel('EstimatedSalary')
axs[1].set_ylabel('Purchased')

# Adjust layout to prevent overlapping of plots.
plt.tight_layout()

# Display the plots.
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

# Create a Logistic Regression model.
model=LogisticRegression()

# Fit the model using 'Age' as the feature and 'Purchased' as the target variable from the 'data' DataFrame.
model.fit(data[['Age']],data.Purchased)
```

```python
# Use the trained Logistic Regression model to predict the 'Purchased' label for a new input feature 'Age' (28).
prediction = model.predict([[28]])
print("Prediction: ",prediction)
```

```python
prediction = model.predict([[58]])
print("Prediction: ",prediction)
```

#**Overfitting** & **Underfitting** & **Best** **fit** **Model**
##**Overfitting** is an undesirable machine learning behavior that occurs when the machine learning model gives accurate predictions for training data but not for test/production data.
##**Underfitting** is when a model has not learned the patterns in the training data well and is unable to generalize well on the test/production data.

##**Best fit Model** is when a model makes the predictions with 0 error, is said to have a good fit on the data. This situation is achievable at a spot between overfitting and underfitting.

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets, using 'Age' as the feature and 'Purchased' as the target variable.
x_train,x_test,y_train,y_test=train_test_split(data[['Age']],data.Purchased, test_size=0.25)

# Fit it using the training data.
model.fit(x_train,y_train)

# Use the trained model to predict the 'Purchased' labels for the test set.
model.predict(x_test)              #We'll always use x_test here
```

```python
#Predict the profit for 50_Startups.csv
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

startup_data = pd.read_csv('/content/50_Startups.csv')

# Assuming 'R&D Spend', 'Administration', 'Marketing Spend' as features, and 'Profit' as the target variable
x = startup_data[['R&D Spend', 'Administration', 'Marketing Spend']]
y = startup_data['Profit']

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)

# Creating a Linear Regression model and fit it using the training data
model = LinearRegression()
model.fit(x_train, y_train)

# Use the trained model to predict the 'Profit' values for the test set
model.predict(x_test)
```
