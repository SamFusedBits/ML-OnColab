```python
#@title Multivariate Regression
import pandas as pd

# Read a CSV file from the specified URL into a DataFrame 'data'. # Create a new DataFrame 'df' to store the data.
url="/content/Multivariate.csv"
data=pd.read_csv(url)
df=pd.DataFrame(data)

# Calculate the median value of the 'Bedroom' column.
median=df.Bedroom.median()

# Replace missing (NaN) values in the 'Bedroom' column with the calculated median value.
df.Bedroom=df.Bedroom.fillna(median) #replace nan value
df
```

```python
#Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns

# Create a linear regression model, fit it to the features 'Area', 'Bedroom', and 'Age' in the DataFrame 'df', with the target variable 'Price' from the 'data' DataFrame.
# Retrieve and display the coefficients for the linear regression model using reg.coef_.

reg=linear_model.LinearRegression()
reg.fit(df[['Area','Bedroom','Age']],data.Price)
reg.coef_
```

```python
# Intercept of the linear regression model
reg.intercept_
```

```python
X = data[['Area', 'Bedroom', 'Age']]
y = data['Price']
# Creating subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Scatter plot for Area vs Price
axs[0].scatter(df['Area'], y)
axs[0].set_title('Area vs Price')
axs[0].set_xlabel('Area')
axs[0].set_ylabel('Price')

# Scatter plot for Bedroom vs Price
axs[1].scatter(df['Bedroom'], y)
axs[1].set_title('Bedroom vs Price')
axs[1].set_xlabel('Bedroom')
axs[1].set_ylabel('Price')

# Scatter plot for Age vs Price
axs[2].scatter(df['Age'], y)
axs[2].set_title('Age vs Price')
axs[2].set_xlabel('Age')
axs[2].set_ylabel('Price')

plt.tight_layout()
plt.show()
```

```python
#Predict price for 3000 sqr ft area, 3 bedrooms, 50 years of age
price=reg.predict([[3000,3,50]])
price[0]
```
