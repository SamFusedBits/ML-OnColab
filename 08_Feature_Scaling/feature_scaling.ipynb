```python
import pandas as pd
import numpy as np
df=pd.read_csv('income.csv')
df
```

```python
#Plotting a scatter plot
from matplotlib import pyplot as plt
plt.scatter(df.Age,df.Income)
```

```python
#Create an object and divide into 3 clusters using n_clusters
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3)
km
```

```python
#Fit the model
y_predict = km.fit_predict(df[['Age','Income']])
y_predict
```

```python
#As we were getting warnings in above code so in order to ignore them we use warnings
import warnings
warnings.filterwarnings('ignore')         #filterwarnings is a function
```

```python
#Adding one column
df['Cluster']=y_predict
df
```

```python
#Separate 3 clusters into 3 dataframes
df1=df[df.Cluster==0]

df2=df[df.Cluster==1]

df3=df[df.Cluster==2]
```

```python
plt.scatter(df1.Age,df1.Income,color='red')
plt.scatter(df2.Age,df2.Income,color='green')
plt.scatter(df3.Age,df3.Income,color='blue')

plt.xlabel("Age")
plt.ylabel("Income")
plt.legend(['df1','df2','df3'])
```

```python
# Feature Scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaler.fit(df[['Income']])
df['Income']=scaler.transform(df[['Income']])

scaler.fit(df[['Age']])
df['Age']=scaler.transform(df[['Age']])
df
```

```python
y_predict = km.fit_predict(df[['Age','Income']])
y_predict
df['Cluster']=y_predict
df
```

```python
df1=df[df.Cluster==0]

df2=df[df.Cluster==1]

df3=df[df.Cluster==2]

plt.scatter(df1.Age,df1.Income,color='red')
plt.scatter(df2.Age,df2.Income,color='green')
plt.scatter(df3.Age,df3.Income,color='blue')

plt.xlabel("Age")
plt.ylabel("Income")

#cluster_centers_ will give x and y coordinates/values of clusters
km.cluster_centers_

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*')
```
